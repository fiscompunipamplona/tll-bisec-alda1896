```python
from pylab import plot, show, grid, xlabel, ylabel, xlim, ylim
from numpy import exp
```

```python
g = 9.8 # m/s^2
m = 68.1 # kg
v = 40. # m/s

def chute(t, c):
    return (g*m)/c * (1-exp(- (c/m) * t)) - v
```

```python
ft = []
t = 10. # s
for i in range(1,30):
    ft.append(chute(t, i))
```

```python
plot(ft, 'o')
xlabel("c")
ylabel("f(c)")
grid()
show()
```

```python
plot(ft, 'o')
xlabel("c")
ylabel("f(c)")

xlim(10,20)
ylim(-10,10)

grid()
show()
```

```python
# Raíz de f(c) usando el método de Bisecón

cl=1
cu=20
cr=(cl+cu)/2

tol=0.01
error=abs((cu-cl)/(cu+cl)) 
err=[]
while error >= tol:
     
    
    if chute(t,cl)*chute(t,cr) < 0:
        cu=cr
    else:
        cl=cr
    cr=(cl+cu)/2  
    
    error=abs((cu-cl)/(cu+cl))
    err.append(error)
       
print ("la raiz es: ", cr)

xlabel("paso")
ylabel("error")

plot(err, 'o-')
```

```python
from numpy import arange
```

```python
ft2 = []
x= []
t = 10. # s
for i in arange(10,20,0.01):
    ft2.append(chute(t, i))
    x.append(i)
```

```python
# Raiz observada por el método grafico en el intervalor [14, 15]
plot(x,ft2, 'o')
xlabel("c")
ylabel("f(c)")

xlim (14,15)
ylim(-2,2)

grid()
show()
```
