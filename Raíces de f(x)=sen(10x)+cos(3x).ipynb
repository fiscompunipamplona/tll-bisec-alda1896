```python

from pylab import plot, show, grid, xlabel, ylabel, xlim, ylim
from numpy import exp
from math import sin, cos
```

```python
def func (x):
    return sin(10*x)+cos(3*x)
```

```python
fx=[]
x=[]
for i in range (0,500):
    fx.append(func(i/100))
    x.append(i/100)
```

```python
plot(x,fx,'o')
xlabel("x")
ylabel("f(x)")
grid()
show()
```

```python
# Se observan las raíces de f(x) en el intervalo [4.2, 4.3]
plot(x,fx, 'o-')
xlabel("x")
ylabel("f(x)")

xlim(4.2,4.3)
ylim(-0.08,0.08)

grid()
show()
```

```python
# se encuentra una raíz de f(x) usando el método de Bisecón

xl=0
xu=5
xr=(xl+xu)/2

tol=0.01
error=error=abs((xu-xl)/(xu+xl))

while error >= tol:
    
    if func(xl)*func(xr) < 0:
        xu=xr
    else:
        xl=xr
    xr=(xl+xu)/2  
    
    error=abs((xu-xl)/(xu+xl)) 
    
print ("la raiz es: ", xr)
        
```
